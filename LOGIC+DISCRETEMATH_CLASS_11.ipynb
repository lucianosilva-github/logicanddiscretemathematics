**MODEL CHECKING WITH CTL**

Computational Tree Logic (CTL) is a logic over computation trees:

Captura de Tela 2023-03-13 às 9.13.34 AM.png

In the model checking we have a formula over te computation trees and we want to retrieve all states in the Kripke structure where the formula is valid.

**CTL MODEL CHECKING IN pyModelChecking**

```python
!pip install pyModelChecking
```

```python
from pyModelChecking import *
K=Kripke(R=[(0,0),(0,1),(1,2),(2,2),(3,3)], L={0: set(['p']), 1:set(['p','q']),3:set(['p'])})
```

```python
from pyModelChecking.CTL import *
phi=AU(True,Or('q',Not(EX('p'))))
print(phi)
```

```python
modelcheck(K,phi)
```

A computer-assisted proof is a mathematical proof that has been at least partially generated by computer.

Most computer-aided proofs to date have been implementations of large proofs-by-exhaustion of a mathematical theorem. The idea is to use a computer program to perform lengthy computations, and to provide a proof that the result of these computations implies the given theorem.
